## Simple function

```python
def f1(x):
    return 3 + (x-2)**2
```

```python
with plt.xkcd(scale=0.6):
    fig, ax = plt.subplots(figsize=(10,6))
    x = np.linspace(-2, 6)
    ax.plot(x, f1(x))
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)');
    ax.set_ylim(bottom=0)
```

```python
def f(x):
    return (x - 2) * (x + 1)**2
```

```python
fig, ax = plt.subplots()
x = np.linspace(-2.2, 2.2)
ax.plot(x, f(x))
ax.grid()
ax.set_xlabel("$x$")
ax.set_ylabel("$f(x)$");
```

```python
# find a local minimum
res = minimize_scalar(f, method='golden')
print(res)
```

```python
def rosenbrock2(x0, x1):

    return 100*(x1 - x0**2)**2 + (1 - x0)**2
```

```python
N = 100

x = np.linspace(-2, 2, N)
y = np.linspace(0, 2, N)
X, Y = np.meshgrid(x, y)
Z = rosenbrock2(X, Y)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.contourf(X, Y, np.log(Z), levels=20, cmap='jet')
ax.set_aspect("equal")
```
